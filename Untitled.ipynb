```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
dataset = pd.read_csv("sales-of-shampoo-over-a-three-ye.csv")
print(dataset.head())
```

```python
dataset = dataset.set_index('Month')

ax = dataset.plot(color='blue')

ax.set_xlabel('Date')

ax.set_ylabel('Sales of Shampoo over a three year period');
```

```python
ax = dataset.plot(kind='hist', bins=50, fontsize=6)

# Set the labels and display the plot
ax.set_xlabel('Date', fontsize=10)
ax.set_ylabel('Sales of Shampoo over a three year period', fontsize=10)
plt.legend(fontsize=10)
plt.show();
```
